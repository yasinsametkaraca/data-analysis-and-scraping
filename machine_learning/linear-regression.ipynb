```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('health.csv')
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
sns.pairplot(df, diag_kind='kde')  # diag_kind: Central density plot. KDE: Density plot
```

Age and visits are positively correlated

```python
X = df[["age"]]
y = df["visits"]
```

```python
y.head()
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
help(train_test_split)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=99)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regr = LinearRegression()
```

```python
regr.fit(X_train, y_train)
```

```python
df["age"].corr(df["visits"])
```

```python
test_pred = regr.predict(X_test)
```

```python
test_pred
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score   # r2_score: Coefficient of determination
```

```python
print("MAE: ", mean_absolute_error(y_test, test_pred)) # MAE: Mean absolute error. Average of absolute residuals. for each data point, it is the difference between the actual value and the predicted value. 4.383435215494492 is the average of the absolute residuals.
print("RMSE: ", np.sqrt(mean_squared_error(y_test, test_pred))) # RMSE: Root mean squared error. Standard deviation of residuals.
print("R2 Score: ", r2_score(y_test, test_pred))  # R2 Score: Coefficient of determination. Model and regression line fit
print("Theta: ", regr.coef_[0])  # theta: curve slope
print("Intercept: ", regr.intercept_) # intercept: curve intercept
```

```python
plt.figure(figsize=(10, 8), dpi=100)
plt.plot(X, regr.predict(X), color='red', linewidth=3)  # X: age, y: visits # regr.predict(X): predicted values
sns.scatterplot(data=df, x="age", y="visits")

plt.xlabel("Age")
plt.ylabel("Visits")
plt.title("Age vs Visits")
```
